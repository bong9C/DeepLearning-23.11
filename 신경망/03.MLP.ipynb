## MLP(Multi Layer Perceptron)

```python
import numpy as np
```

```python
def AND_gate(x):                # x는 튜플
    weight = np.array([1,1])
    bias = -1
    sum = np.dot(weight, x) + bias
    return 1 if sum > 0 else 0
```

```python
def OR_gate(x):
    weight = np.array([2,2])
    bias = -1
    sum = np.dot(weight, x) + bias
    return 1 if sum > 0 else 0
```

```python
def NAND_gate(x):
    weight = np.array([-2,-2])
    bias = 3
    sum = np.dot(weight, x) + bias
    return 1 if sum > 0 else 0
```

- MLP

- MLP

```python
h1, h2 = [], [] 
for x in [(0,0), (0,1), (1,0), (1,1)]:
    h1.append(NAND_gate(x))
    h2.append(OR_gate(x))
h1,h2
```

```python
for x in zip(h1, h2):
    print(AND_gate(x))
```

- XOR gate

```python
def XOR_gate(x):
    return AND_gate([NAND_gate(x), OR_gate(x)])
```

```python
for x in [(0,0),(0,1),(1,0),(1,1)]:
    print(f'{x} :\t{XOR_gate(x)}')
```
