## 활성화(Activation) 함수

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(-5, 5, 501)
```

- heaviside step function 

```python
def step(x):
    y = np.zeros_like(x)
    y[x > 0] = 1            # x > 0 이면 1, x <= 0 이면 0
    return y
```

```python
plt.figure(figsize=(5,3))
plt.plot(xs, step(xs))
plt.grid(), plt.title('heaviside step function')
plt.show()
```

- hyperbolic tangent

```python
plt.figure(figsize=(5,3))
plt.plot(xs, np.tanh(xs))
plt.grid(),plt.title('hyperbolic tangent function')
plt.show()
```

- relu(rectifed linesr)

```python
def relu(x):
    return np.maximum(0, x)         # relu 이용시: maximum 
```

```python
plt.figure(figsize=(5,3))
plt.plot(xs, relu(xs))
plt.grid(),plt.title('relu function')
plt.show()
```

- leaky relu

```python
def leaky_relu(x):
    return np.maximum(0.1*x,x)
```

```python
plt.figure(figsize=(5,3))
plt.plot(xs, leaky_relu(xs))
plt.grid(),plt.title('leaky relu function')
plt.show()
```

- softplus

```python
def softpius(x):
    return np.log(1 + np.exp(x))
```

```python
plt.figure(figsize=(5,3))
plt.plot(xs, softpius(xs))
plt.plot(xs, relu(xs))
plt.grid(),plt.title('softpius function')
plt.show()
```

- softmax

```python
def softmax(x):
    return np.exp(x) /  np.sum(np.exp(x))
```

```python
yhat = np.array([3.0, 1.5, 3.0])
softmax(yhat)
```
